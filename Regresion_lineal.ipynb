```python
#Regresión Lineal 5  Grados

#Datos

x = [0.33,0.4,0.6,0.73,0.79,0.93,1.14,1.18,1.23,1.33]

y = [26,36,46,56,66,76,86,96,106,116]
```

Formulas utilizadas para realizar la regresión lineal:

1. Ecuación de la recta:

$$
    y=mx+b
$$

2. Ecuación de la pendiente:

$$

    m = \frac{\Sigma x \Sigma y - n \Sigma (xy)}{(\Sigma x)^2 - n \Sigma x^2}

$$

3. Ecuación del intercepto (b):

$$

    b = \bar{y} - m\bar{x}
$$

4. Ecuación para hallar $(\sigma_{x})$:
$$

    \sigma_{x} = \sqrt{\frac{\Sigma (x^2)}{n}-\bar{x}^2}
$$

5. Ecuación para hallar $(\sigma_{y})$:
$$

    \sigma_{x} = \sqrt{\frac{\Sigma (y^2)}{n}-\bar{y}^2}
$$
6. Ecuación para hallar $(\sigma_{xy})$:
$$
    \sigma_{xy}= \frac{\Sigma (xy)}{n}-\bar{x}\bar{y}
$$
7. Ecuación para hallar $R^2$:
$$
    R = \frac{\sigma_{xy}}{\sigma_{x}\sigma_{y}}
$$
8. Ecuación para velocidad promedio:
$$
    v = \frac{2(x-x_{0})}{t}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python

k=0
y0 = 16
y5 = []
for i in range(len(x)):
    formula = (2*(y[k]-y0))/x[k]
    y5.append(formula)
    k+=1
for i in range(len(y5)):
    print(y5[i]) #Obtener los valores de la velocidad promedio con respecto a tiempo.
```

```python
n = len(x)
x = np.array(x)
y = np.array(y5)
sumx = sum(x) #Sumatoria de x
sumy = sum(y) #Sumatoria de y
sumx2 = sum(x**2) #Sumatoria de x^2
sumy2 = sum(y**2) #Sumatoria de y^2
sumxy = sum(x*y) #Sumatoria de x*y
promx = sumx/n #Promedio de x
promy = sumy/n #Promedio de y
```

```python
m = (sumx*sumy - n*sumxy)/(sumx**2 - n*sumx2) #Pendiente
b = promy - m*promx #Intercepto
m, b
```

```python
sigmax = np.sqrt(sumx2/n - promx**2) #Desviación estándar de x
sigmay = np.sqrt(sumy2/n - promy**2) #Desviación estándar de y
sigmaxy = sumxy/n - promx*promy #Covarianza
R2 = (sigmaxy/(sigmax*sigmay))**2 #Coeficiente de correlación
```

```python
plt.plot(x, y, 'o', label = 'Datos', color = 'orange')
plt.plot(x, m*x + b, label = ' y = {:.2f}x + {:.2f}\n R2 = {:.2f}'.format(round(m, 2), round(b, 2), round(R2, 2)))
plt.xlabel('Tiempos (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Regresión Lineal 5 Grados')
plt.grid()
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```

```python
#Regresión Lineal 10  Grados

#Datos

x_10 = [0.16,0.25,0.33,0.43,0.53,0.63,0.68,0.76,0.89,0.88]

y_10 = [26,36,46,56,66,76,86,96,106,116]
```

```python
plt.plot(x_10, y_10, 'o', label = 'Datos', color = 'orange')
plt.plot(x_10, m10*x_10 + b10, label = ' y = {:.2f}x + {:.2f}\n R2 = {:.2f}'.format(round(m10, 2), round(b10, 2), round(R2_10, 2)))
plt.xlabel('Tiempos (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Regresión Lineal 10 Grados')
plt.grid()
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```
