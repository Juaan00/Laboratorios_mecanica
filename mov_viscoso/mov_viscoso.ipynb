```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import stats
```

```python
datos = pd.read_excel('datos_mov_viscoso.xlsx')
datos
```

```python
x = datos['x']
t_10 = datos['t_10']
t_20 = datos['t_20']
t_30 = datos['t_30']
t_40 = datos['t_40']
t_50 = datos['t_50']
```

```python

def regresion_1(t):
  slope, intercept, r, p, std_err = stats.linregress(x, t)
  def myfunc(x):
    return slope * x + intercept
  
  mymodel = list(map(myfunc, x))
  print('slope:', slope)
  print('intercept:', intercept)
  print('r:', r)
  print('p:', p)
  print('std_err:', std_err)
  return mymodel

t_reg_10 = regresion_1(t_10)
t_reg_20 = regresion_1(t_20)
t_reg_30 = regresion_1(t_30)
t_reg_40 = regresion_1(t_40)
t_reg_50 = regresion_1(t_50)
```

```python
fig = plt.figure(figsize=(10, 5), layout='tight')
fig.suptitle('Movimiento viscoso (Datos experimentales)', fontsize=16)
spec = fig.add_gridspec(2,3)
ax1 = fig.add_subplot(spec[0, 0])
ax1.plot(x, t_reg_10,linestyle='--', label='y10 = 0.34(x)+0.17 \n R10 = 0.99', color='black')
ax1.plot(x,t_10, 'o', label='h = 10 cm',color='blue')
ax1.set_yticks(np.arange(0, 30, 5))
ax1.set_xticks(np.arange(0, 95, 10))
ax1.set_xlabel('Distancia (cm)')
ax1.set_ylabel('Tiempo (s)')
ax1.grid(True)
ax2 = fig.add_subplot(spec[0, 1])
ax2.plot(x,t_reg_20,linestyle='--', label='y20 = 0.21(x)+0.22 \n R20 = 0.99', color='black')
ax2.plot(x,t_20, 'o', label='h = 20 cm',color='violet')
ax2.set_yticks(np.arange(0, 30, 5))
ax2.set_xticks(np.arange(0, 95, 10))
ax2.set_xlabel('Distancia (cm)')
ax2.set_ylabel('Tiempo (s)')
ax2.grid(True)
ax3 = fig.add_subplot(spec[0, 2])
ax3.plot(x,t_reg_30,linestyle='--', label='y30 = 0.17(x)+0.07 \n R30 = 0.99', color='black')
ax3.plot(x,t_30, 'o', label='h = 30 cm', color='green')
ax3.set_yticks(np.arange(0, 30, 5))
ax3.set_xticks(np.arange(0, 95, 10))
ax3.set_xlabel('Distancia (cm)')
ax3.set_ylabel('Tiempo (s)')
ax3.grid(True)
ax4 = fig.add_subplot(spec[1, 0])
ax4.plot(x,t_reg_40,linestyle='--', label='y40 = 0.15(x)+0.07 \n R40 = 0.99', color='black')
ax4.plot(x,t_40, 'o', label='h = 40 cm', color='red')
ax4.set_yticks(np.arange(0, 30, 5))
ax4.set_xticks(np.arange(0, 95, 10))
ax4.set_xlabel('Distancia (cm)')
ax4.set_ylabel('Tiempo (s)')
ax4.grid(True)
ax5 = fig.add_subplot(spec[1, 1])
ax5.plot(x,t_reg_50,linestyle='--', label='y50 = 0.14(x)+0.02 \n R50 = 0.99', color='black')
ax5.plot(x,t_50, 'o', label='h = 50 cm', color='orange')
ax5.set_yticks(np.arange(0, 30, 5))
ax5.set_xticks(np.arange(0, 95, 10))
ax5.set_xlabel('Distancia (cm)')
ax5.set_ylabel('Tiempo (s)')
ax5.grid(True)
ax6 = fig.add_subfigure(spec[1,2])
fig.legend(loc='lower center',bbox_to_anchor=(0.5, -0.15), ncol=5,shadow=True, fancybox=True)
plt.savefig('datos_experimentales_viscoso.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
