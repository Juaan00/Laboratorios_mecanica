```python
l = 55

m = [66,54,43] #31

a = [5,10,15]

m_1_a_5 = np.array([8.59,8.60,8.62])
m1_a_10 = np.array([8.79,8.78,8.75])
m1_a_15 = np.array([8.85,8.93,8.89])

m2_a_5 = np.array([8.67,8.66,8.64])
m2_a_10 = np.array([8.82,8.73,8.78])
m2_a_15 = np.array([8.72,8.74,8.72])

m3_a_5 = np.array([8.70,8.73,8.72])
m3_a_10 = np.array([8.74,8.67,8.79])
m3_a_15 = np.array([8.85,8.85,8.82])

m1_a_5_d = [(m_1_a_5[i]/6) for i in range(3)]
m1_a_10_d = [(m1_a_10[i]/6) for i in range(3)]
m1_a_15_d = [(m1_a_15[i]/6) for i in range(3)]

m2_a_5_d = [(m2_a_5[i]/6) for i in range(3)]
m2_a_10_d = [(m2_a_10[i]/6) for i in range(3)]
m2_a_15_d = [(m2_a_15[i]/6) for i in range(3)]

m3_a_5_d = [(m3_a_5[i]/6) for i in range(3)]
m3_a_10_d = [(m3_a_10[i]/6) for i in range(3)]
m3_a_15_d = [(m3_a_15[i]/6) for i in range(3)]

tabla_valores = pd.DataFrame({
    'm1 5 grados': m1_a_5_d,
    'm2 5 grados': m2_a_5_d,
    'm3 5 grados': m3_a_5_d,
    'm1 10 grados': m1_a_10_d,
    'm2 10 grados': m2_a_10_d,
    'm3 10 grados': m3_a_10_d,
    'm1 15 grados': m1_a_15_d,
    'm2 15 grados': m2_a_15_d,
    'm3 15 grados': m3_a_15_d
})
tabla_valores

```

```python
def gravedad(l,t):
    return l*(((2*np.pi)**2)/(t**2))

grav_m1_5 = [gravedad(l,m1_a_5_d[i]) for i in range(3)]
grav_m2_5 = [gravedad(l,m2_a_5_d[i]) for i in range(3)]
grav_m3_5 = [gravedad(l,m3_a_5_d[i]) for i in range(3)]

grav_m1_10 = [gravedad(l,m1_a_10_d[i]) for i in range(3)]
grav_m2_10 = [gravedad(l,m2_a_10_d[i]) for i in range(3)]
grav_m3_10 = [gravedad(l,m3_a_10_d[i]) for i in range(3)]

grav_m1_15 = [gravedad(l,m1_a_15_d[i]) for i in range(3)]
grav_m2_15 = [gravedad(l,m2_a_15_d[i]) for i in range(3)]
grav_m3_15 = [gravedad(l,m3_a_15_d[i]) for i in range(3)]

tabla_gravedad = pd.DataFrame({
    'm1 5 grados': grav_m1_5,
    'm2 5 grados': grav_m2_5,
    'm3 5 grados': grav_m3_5,
    'm1 10 grados': grav_m1_10,
    'm2 10 grados': grav_m2_10,
    'm3 10 grados': grav_m3_10,
    'm1 15 grados': grav_m1_15,
    'm2 15 grados': grav_m2_15,
    'm3 15 grados': grav_m3_15
})
tabla_gravedad
```

```python
def incertidumbre (l,t):
    def incer_l(l):
        return (2*np.pi)/l
    def incer_t(l,t):
        return (l*2*np.pi)/(t**2)
    return np.sqrt(((incer_l(l)*0.1)**2)+((incer_t(l,t)*0.016)**2))

incer_gravedad_m1_5 = [incertidumbre(l,m1_a_5_d[i]) for i in range(3)]
incer_gravedad_m2_5 = [incertidumbre(l,m2_a_5_d[i]) for i in range(3)]
incer_gravedad_m3_5 = [incertidumbre(l,m3_a_5_d[i]) for i in range(3)]

incer_gravedad_m1_10 = [incertidumbre(l,m1_a_10_d[i]) for i in range(3)]
incer_gravedad_m2_10 = [incertidumbre(l,m2_a_10_d[i]) for i in range(3)]
incer_gravedad_m3_10 = [incertidumbre(l,m3_a_10_d[i]) for i in range(3)]

incer_gravedad_m1_15 = [incertidumbre(l,m1_a_15_d[i]) for i in range(3)]
incer_gravedad_m2_15 = [incertidumbre(l,m2_a_15_d[i]) for i in range(3)]
incer_gravedad_m3_15 = [incertidumbre(l,m3_a_15_d[i]) for i in range(3)]

tabla_incertidumbre = pd.DataFrame({
    'm1 5 grados': incer_gravedad_m1_5,
    'm2 5 grados': incer_gravedad_m2_5,
    'm3 5 grados': incer_gravedad_m3_5,
    'm1 10 grados': incer_gravedad_m1_10,
    'm2 10 grados': incer_gravedad_m2_10,
    'm3 10 grados': incer_gravedad_m3_10,
    'm1 15 grados': incer_gravedad_m1_15,
    'm2 15 grados': incer_gravedad_m2_15,
    'm3 15 grados': incer_gravedad_m3_15
})
tabla_incertidumbre
```
